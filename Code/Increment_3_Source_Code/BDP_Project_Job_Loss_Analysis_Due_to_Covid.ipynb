```python
import pandas as pd
```

```python
df=pd.read_csv('C:\\Users\\muppa\\Big_Data Project_groupby_Output.csv')
```

```python
df['Count'].sum()
```

```python
df_temp=df.groupby('Sector')['Count'].sum()
```

```python
df_temp.values
```

## Job loss distribution Sector wise in Pie Chart

```python
import matplotlib.pyplot as plt

my_labels = df_temp.index
my_data = df_temp.values
plt.pie(my_data,labels=my_labels,autopct='%1.1f%%')
plt.title('Job loss Distribution Sector wise \n \n')
plt.axis('equal')
plt.show()
```

## Job loss distribution Sector wise in Pie Chart

```python
df_temp=df.groupby('Country')['Count'].sum()
```

```python
import matplotlib.pyplot as plt

my_labels = df_temp.index
my_data = df_temp.values
plt.pie(my_data,labels=my_labels,autopct='%1.1f%%')
plt.title('Job loss Distribution Country wise \n \n')
plt.axis('equal')
plt.show()
```

## Job loss in US sector wise

```python
df1=df[df['Country']=='US']
fig = plt.figure()
ax = fig.add_axes([0,0,3,3])
sector = df1.Sector
count = df1.Count
ax.bar(sector,count)
plt.show()
```
